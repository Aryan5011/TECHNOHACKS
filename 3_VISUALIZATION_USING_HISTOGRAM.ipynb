```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('iris data.csv')
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(5, 5))
sns.countplot(x='Species',data=data)
plt.title("Species Distribution")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

```python
print(data['Species'].value_counts())
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(10, 10))
sns.scatterplot(x="SepalLengthCm", y="PetalLengthCm", hue="Species", data=data, palette="Blues", s=100)
plt.title("Sepal Length vs Petal Length by Species")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(10, 10))
sns.scatterplot(x="SepalWidthCm", y="PetalWidthCm", hue="Species", data=data, palette="Set1", s=100)
plt.title("Sepal Width vs Petal Width by Species")
plt.xlabel("Sepal Width")
plt.ylabel("Petal Width")
plt.legend(title="Species")
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.boxplot(x='Species',y='SepalLengthCm',data=data ,palette='Set1')
```

```python
sns.pairplot(data, hue='Species',palette='Set1')
```

```python
plt.figure(figsize=(8, 8))
sns.scatterplot(x='PetalLengthCm',y='PetalWidthCm',data=data,hue='Species', palette= 'Paired',s=100)
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.show()
```

```python
plt.figure(figsize=(8, 8))
sns.scatterplot(x='SepalLengthCm',y='SepalWidthCm',data=data,hue='Species',s=100)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

```python
correlation = data.corr()
plt.figure(figsize=(8,6))
sns.heatmap(correlation, annot=True, cmap='rocket', fmt=".2f")
plt.show()
```
